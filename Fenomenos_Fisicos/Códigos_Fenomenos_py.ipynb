```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, limit, sympify

# Função para calcular o limite e plotar o gráfico
def calcular_limite(expressao, variavel, valor):
    # Definindo a variável simbólica
    x = symbols(variavel)

    # Convertendo a string da expressão para uma expressão simbólica
    expressao_simbolica = sympify(expressao)

    # Plotando o gráfico da função
    x_vals = np.linspace(valor - 2, valor + 2, 400)
    y_vals = [expressao_simbolica.subs(x, val) for val in x_vals]
    plt.plot(x_vals, y_vals, label='Função')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gráfico da função')
    plt.axvline(x=valor, color='r', linestyle='--', label=f'Limite em x={valor}')

    # Calculando o limite
    limite = limit(expressao_simbolica, x, valor)

    # Exibindo o valor do limite
    print("Limite:", limite)

    # Plotando o ponto do limite
    plt.scatter(valor, limite, color='green', label=f'Limite = {limite}')

    plt.legend()
    plt.grid(True)
    plt.show()

    return expressao_simbolica, limite

# Definindo a variável simbólica
x = symbols('x')

# Input da expressão e variável
variavel = 'x'
expressao = (x**4-1)/(x**-4)
valor = 2

# Calculando o limite e plotando o gráfico
calcular_limite(expressao, variavel, valor)
#(x**3-8)/(x**2-4)
```

```python
from sympy import symbols, sympify
import numpy as np
from IPython.display import display, Math

# Definindo a variável simbólica
x = symbols('x')

# Input da expressão como uma string
variavel = 'x'
expressao = (x**4-1)/(x**-4)

# Convertendo a string da expressão para uma expressão simbólica
f = sympify(expressao)

# Input do valor do ponto desejado
ponto_desejado = float(-4)

def criar_pontos_aproximacao(valor_limite, distancia=[0.1, 0.01, 0.001, 0.0001, 0.00001]):
    pontos_esquerda = [valor_limite - d for d in distancia]
    pontos_direita = [valor_limite + d for d in distancia]
    return pontos_esquerda, pontos_direita

# Reaplicando a função para criar os pontos de aproximação
pontos_esquerda, pontos_direita = criar_pontos_aproximacao(ponto_desejado)

# Recalculando os valores de f(x) para os pontos à esquerda e à direita
valores_esquerda = [f.subs(x, ponto).evalf() for ponto in pontos_esquerda]
valores_direita = [f.subs(x, ponto).evalf() for ponto in pontos_direita]

# Preparando a saída em formato de tabela novamente
tabela_latex = "\\begin{array}{|c|c|c|c|}\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\text{Aprox. à Esquerda} & \\text{Valor} & \\text{Aprox. à Direita} & \\text{Valor} \\\\\n"
tabela_latex += "\\hline\n"
for i in range(len(pontos_esquerda)):
    tabela_latex += f"{pontos_esquerda[i]:.5f} & {valores_esquerda[i]:.5f} & {pontos_direita[i]:.5f} & {valores_direita[i]:.5f} \\\\\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\end{array}"

display(Math(tabela_latex))
```
